```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
INDEX_FILE = "D:\\petrtsv\\projects\\ds\\pytorch-sessions\\index.csv"

records = pd.read_csv(INDEX_FILE, index_col=0)
records = records[records['record'] == 1]

records
```

```python
current = records[
    ((records.index >= 233) & (records.index <= 240)) |
    ((records.index >= 241) & (records.index <= 244)) |
    ((records.index >= 249) & (records.index <= 256)) |
    ((records.index >= 257) & (records.index <= 260)) |
    
    
    ((records.index >= 265) & (records.index <= 268)) |
    ((records.index >= 269) & (records.index <= 272)) |
    ((records.index >= 273) & (records.index <= 276)) |
    ((records.index >= 277) & (records.index <= 280)) 
]

current
```

```python
COLUMNS = ['dataset', 'test_accuracy',
           'reduced', 'test_balanced_accuracy', 'epochs', 'batch_size', 'frozen_feature_extractor', 'augment_prob']

current = current[COLUMNS]
current
```

```python
cifar_sc_records = current[(
    current['dataset'] == 'CIFAR10') &
    (current.reduced < 1) &
    (current.frozen_feature_extractor == 1)].sort_values(by='reduced', ascending=True)

cifar_sc_records_not_fixed = current[(
    current['dataset'] == 'CIFAR10') &
    (current.reduced < 1) &
    (current.frozen_feature_extractor == 0)].sort_values(by='reduced', ascending=True)

x_labels = ["%d%% (%d)" % (int((1 - label) * 100), int(50000*(1-label)))
            for label in cifar_sc_records['reduced']]
h1 = cifar_sc_records['test_accuracy']
h2 = cifar_sc_records['test_balanced_accuracy']

h1_nf = cifar_sc_records_not_fixed['test_accuracy']
h2_nf = cifar_sc_records_not_fixed['test_balanced_accuracy']

width = 0.25

pad = 0.75

x1 = np.arange(len(h1)) * pad
x2 = [x + width for x in x1]

plt.figure(figsize=(15, 10))

plt.bar(x1, h1_nf, width=width / 3, label='Accuracy')
plt.bar(x2, h2_nf, width=width / 3, label='Balanced Accuracy')

plt.bar(x1, h1, width=width, label='Accuracy (fixed feature extractor)')
plt.bar(x2, h2, width=width, label='Balanced Accuracy (fixed feature extractor)')

plt.xlabel('Dataset: percentage of samples (samples number)', fontweight='bold')
plt.ylabel('Value', fontweight='bold')
plt.xticks([x * pad + width / 2 for x in range(len(h1))], x_labels, rotation=-45)
plt.yticks(np.arange(0, 1.05, step=0.05))


plt.title("CIFAR10")

plt.legend()
plt.show()
```
