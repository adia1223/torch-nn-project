```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
INDEX_FILE = "D:\\petrtsv\\projects\\ds\\pytorch-sessions\\index.csv"

records = pd.read_csv(INDEX_FILE, index_col=0)
records = records[records['record'] == 1]

records.tail(50)
```

```python
current = records[
    
    # baseline
    ((records.index >= 265) & (records.index <= 268)) | # aug 50%, cifar, (5, 1) samples, (1, 0) frozen feature extractor
#     ((records.index >= 269) & (records.index <= 272)) | # aug 0%, cifar, (5, 1) samples, (1, 0) frozen feature extractor
    ((records.index >= 273) & (records.index <= 276)) | # aug 50%, gtsrb, (5, 1) samples, (1, 0) frozen feature extractor
#     ((records.index >= 277) & (records.index <= 280)) | # aug 0%, gtsrb, (5, 1) samples, (1, 0) frozen feature extractor
    
    # baseline++ (cosine based)
    ((records.index >= 298) & (records.index <= 301)) | # aug 50%, cifar, (5, 1) samples, (1, 0) frozen feature extractor
#     ((records.index >= 269) & (records.index <= 272)) | # aug 0%, cifar, (5, 1) samples, (1, 0) frozen feature extractor
    ((records.index >= 302) & (records.index <= 305)) | # aug 50%, gtsrb, (5, 1) samples, (1, 0) frozen feature extractor
#     ((records.index >= 277) & (records.index <= 280)) | # aug 0%, gtsrb, (5, 1) samples, (1, 0) frozen feature extractor
    
    (records.index == -282)
]
```

```python
COLUMNS = ['dataset', 'test_accuracy', 'backbone',
           'reduced', 'test_balanced_accuracy', 'frozen_feature_extractor']

scores = current[COLUMNS]
```

# Models scores

```python
for dataset in ('CIFAR10', 'GTSRB'):
    print("Scores on %s" % dataset)
    for samples in (5, 1):
        print("\t %d samples per class" % samples)
        for model in ('ResNet18', 'ResNet18Cosine'):
            print("\t\t Scores for model %s" % model)
            
            print()
            print("\t\t\t Frozen backbone\tAccuracy\tBalanced accuracy")
            for frozen in (0.0, 1.0):
                record = scores[
                    (scores.dataset == dataset) &
                    (scores.reduced == samples) & 
                    (scores.backbone == model) &
                    (scores.frozen_feature_extractor == frozen)
                ]
                accuracy = max(record.test_accuracy)
                balanced_accuracy = max(record.test_balanced_accuracy)
                print("\t\t\t %d\t\t\t%.3f\t\t\t%.3f" % (frozen, accuracy, balanced_accuracy))
            print()
```
